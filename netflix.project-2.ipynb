```python
df = df.drop(columns=['director'])
```

```python
df.head()
```

```python
df1 = df1.drop(columns=['director'])
```

```python
df1.head()
```

```python
df1['date_added'] = pd.to_datetime(df1['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
df1['date_added'].head()
```

```python
df1['year_added'] = df1['date_added'].dt.year
df1['month_added'] = df1['date_added'].dt.month
```

```python
df1.head()
```

```python
df1['year_added'] = df1['year_added'].astype('Int64')
df1['month_added'] = df1['month_added'].astype('Int64')
```

```python
df1.head()
```

```python
yearly = df1.groupby('year_added')['show_id'].count()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
yearly.plot(kind='bar', color='skyblue')
plt.title("Количество фильмов и сериалов, добавленных на Netflix по годам", fontsize=14)
plt.xlabel("Год добавления")
plt.ylabel("Количество")
plt.show()
```

```python
heatmap_data = df1.pivot_table(index='year_added', 
                               columns='month_added', 
                               values='show_id', 
                               aggfunc='count')

import seaborn as sns

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlOrRd", annot=True, fmt="g")
plt.title("Добавления контента Netflix по годам и месяцам", fontsize=14)
plt.show()
```

```python
type_year = df1.groupby(['year_added', 'type'])['show_id'].count().unstack()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
type_year.plot(kind='bar', stacked=False, figsize=(14,6))
plt.title("Movies vs TV Shows по годам добавления", fontsize=14)
plt.xlabel("Год")
plt.ylabel("Количество")
plt.legend(title="Тип контента")
plt.show()
```

```python
plt.figure(figsize=(12,6))
type_year.plot(kind='line', marker='o', figsize=(14,7))
plt.title("Динамика добавлений Movies и TV Shows по годам", fontsize=14)
plt.xlabel("Год")
plt.ylabel("Количество")
plt.legend(title="Тип контента")
plt.show()
```

```python
movies = df1[df1['type'] == 'Movie'].copy()
movies['quarter'] = movies['date_added'].dt.quarter

heatmap_data = movies.pivot_table(
    index='year_added',
    columns='quarter',
    values='show_id',
    aggfunc='count'
)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlOrRd", annot=True, fmt="g")
plt.title("Количество фильмов, добавленных на Netflix (по годам и кварталам)", fontsize=14)
plt.xlabel("Квартал")
plt.ylabel("Год")
plt.show()
```

```python
movies = df1[df1['type'] == 'TV Show'].copy()
movies['quarter'] = movies['date_added'].dt.quarter

heatmap_data = movies.pivot_table(
    index='year_added',
    columns='quarter',
    values='show_id',
    aggfunc='count'
)

plt.figure(figsize=(10,5))
sns.heatmap(heatmap_data, cmap="YlOrRd", annot=True, fmt="g")
plt.title("Количество TV Show, добавленных на Netflix (по годам и кварталам)", fontsize=10)
plt.xlabel("Квартал", fontsize=8)
plt.ylabel("Год", fontsize=8)

plt.xticks(fontsize=10)
plt.yticks(fontsize=8)

plt.show()
```

```python
genres_per_year = (
    df1
    .dropna(subset=['listed_in'])
    .assign(genre=df1['listed_in'].str.split(','))
    .explode('genre')
)

genres_per_year['genre'] = genres_per_year['genre'].str.strip()

genre_year = (
    genres_per_year
    .groupby(['year_added','genre'])['show_id']
    .count()
    .reset_index()
)

pivot_genre_year = genre_year.pivot(
    index='genre',
    columns='year_added',
    values='show_id'
).fillna(0).astype(int)

plt.figure(figsize=(14,10))
sns.heatmap(
    pivot_genre_year,
    cmap="YlOrRd",
    cbar=True,
    linewidths=1
)

plt.title("Динамика жанров Netflix по годам", fontsize=16)
plt.xlabel("Год")
plt.ylabel("Жанр")
plt.show()
```
